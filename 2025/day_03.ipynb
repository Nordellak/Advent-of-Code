## Part 1

Puzzle input:

```python
with open("input_files/day_03.txt") as input_file:
    input = input_file.read()
```

Parse the input:

```python
battery_banks = [list(map(int, list(line))) for line in input.split("\n")]
```

```python
total_joltage = 0
for bank in battery_banks:
    max_joltage = max(bank)
    first_battery = bank.index(max_joltage)
    if first_battery == len(bank) - 1:
        max_joltage = max(bank[:-1])
        first_battery = bank.index(max_joltage)
    second_max_joltage = max(bank[first_battery + 1:])
    total_joltage += int(str(max_joltage) + str(second_max_joltage))
    
print(f"The total output joltage is {total_joltage}.")
```

## Part 2

```python
total_joltage = 0
for bank in battery_banks:
    bank_max_joltage_str = ""
    remaining_bank = bank.copy()
    for battery_num in range(12, 0, -1):
        available_batteries = remaining_bank[:-battery_num + 1]
        if len(available_batteries) == 0:
            max_joltage = max(remaining_bank)
        else:
            max_joltage = max(available_batteries)
        index_battery = remaining_bank.index(max_joltage)
        remaining_bank = remaining_bank[index_battery + 1:]
        bank_max_joltage_str += str(max_joltage)
        
    total_joltage += int(bank_max_joltage_str)

print(f"The new total output joltage is {total_joltage}.")
```

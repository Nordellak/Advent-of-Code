## Part 1

Puzzle input:

```python
with open("input_files/day_05.txt") as input_file:
    input = input_file.read()
```

Parse the input:

```python
fresh_ranges_raw, available_raw = input.split("\n\n")
fresh_ranges = [list(map(int, line.split("-"))) for line in fresh_ranges_raw.split("\n")]
available = [int(aid) for aid in available_raw.split("\n")]
```

```python
fresh_ingredients = 0
for ingredient_id in available:
    for range in fresh_ranges:
        if ingredient_id >= range[0] and ingredient_id <= range[1]:
            fresh_ingredients += 1
            break
print(f"There are {fresh_ingredients} fresh ingredients.")
```

## Part 2

The following function wasn't necessary after all, it was part of my initial attempt. I was calculating for every fresh range if there was any overlapping IDs with the previously looked at fresh ranges, and subtracting the number to not account for fresh IDs more than once. This worked for the example case, but didn't work with the actual input, and after a while I realized that the issue was that I was removing too many IDs, as there are a lot of overlapping ranges, and I was looking only at one at a time.

```python
# Unused function
def overlapping_ids(range, looked_range):
    if range[0] < looked_range[0] and range[1] < looked_range[0]:
        return 0
    if range[0] > looked_range[1] and range[1] > looked_range[1]:
        return 0
    if range[0] >= looked_range[0] and range[1] <= looked_range[1]:
        return range[1] - range[0] + 1
    if range[0] <= looked_range[0] and range[1] >= looked_range[1]:
        return looked_range[1] - looked_range[0] + 1
    if range[0] < looked_range[0] and range[1] >= looked_range[0] and range[1] <= looked_range[1]:
        return range[1] - looked_range[0] + 1
    if range[0] >= looked_range[0] and range[0] <= looked_range[1] and range[1] > looked_range[1]:
        return looked_range[1] - range[0] + 1
    return None
```

Finally I thought of creating non overlapping ranges, but couldn't figure out how to do that easily. It was easier for me to create a non overlapping range of **non** fresh ingredients. I started with a range encompassing all the possible values, and began chipping away the fresh ones. Once I had the non overlapping list of non fresh ingredients, it was just a matter of subtracting the total amount of non fresh ingredients to the total amount of fresh ingredients (`max_id - min_id`).

The many if statements could probably be reduced, but this is what I had at the time of solving it. ðŸ™‚

```python
min_id = fresh_ranges[0][0]
max_id = fresh_ranges[0][1]
for range in fresh_ranges:
    if range[0] < min_id:
        min_id = range[0]
    if range[1] > max_id:
        max_id = range[1]

not_fresh_ranges = [[min_id, max_id]]
for fresh_range in fresh_ranges:
    not_fresh_ranges_new = not_fresh_ranges.copy()
    for not_fresh_range in not_fresh_ranges:
        if fresh_range[0] == not_fresh_range[0] and fresh_range[1] == not_fresh_range[1]:
            not_fresh_ranges_new.remove(not_fresh_range)
        elif fresh_range[0] == not_fresh_range[0] and fresh_range[1] < not_fresh_range[1]:
            not_fresh_ranges_new.remove(not_fresh_range)
            not_fresh_ranges_new.append([fresh_range[1] + 1, not_fresh_range[1]])
        elif fresh_range[0] > not_fresh_range[0] and fresh_range[1] == not_fresh_range[1]:
            not_fresh_ranges_new.remove(not_fresh_range)
            not_fresh_ranges_new.append([not_fresh_range[0], fresh_range[0] - 1])
        elif fresh_range[0] > not_fresh_range[0] and fresh_range[1] < not_fresh_range[1]:
            not_fresh_ranges_new.remove(not_fresh_range)
            not_fresh_ranges_new.append([not_fresh_range[0], fresh_range[0] - 1])
            not_fresh_ranges_new.append([fresh_range[1] + 1, not_fresh_range[1]])
        elif fresh_range[0] <= not_fresh_range[0] and fresh_range[1] >= not_fresh_range[1]:
            not_fresh_ranges_new.remove(not_fresh_range)
        elif fresh_range[0] < not_fresh_range[0] and fresh_range[1] >= not_fresh_range[0] and fresh_range[1] < not_fresh_range[1]:
            not_fresh_ranges_new.remove(not_fresh_range)
            not_fresh_ranges_new.append([fresh_range[1] + 1, not_fresh_range[1]])
        elif fresh_range[0] > not_fresh_range[0] and fresh_range[0] <= not_fresh_range[1] and fresh_range[1] > not_fresh_range[1]:
            not_fresh_ranges_new.remove(not_fresh_range)
            not_fresh_ranges_new.append([not_fresh_range[0], fresh_range[0] - 1])
    not_fresh_ranges = not_fresh_ranges_new.copy()

fresh_ingredients = max_id - min_id + 1
for range in not_fresh_ranges:
    fresh_ingredients -= range[1] - range[0] + 1
print(f"The total amount of fresh ingredients is {fresh_ingredients}.")
```

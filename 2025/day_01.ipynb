## Part 1

Puzzle input:

```python
with open("input_files/day_01.txt") as input_file:
    input = input_file.read()
```

Test input:

```python
# # Comment this cell to use the puzzle input instead of the test input
# input = """L68
# L30
# R48
# L5
# R60
# L55
# L1
# L99
# R14
# L82"""
```

Parse the input:

```python
input_split = input.split("\n")
dial_rotations = [int(rotation[1:]) if rotation[:1] == "R" else -int(rotation[1:]) for rotation in input_split]

starting_position = 50
```

```python
password = 0
previous_position = starting_position
for rotation in dial_rotations:
    position = (previous_position + rotation) % 100
    
    if position == 0:
        password += 1
        
    previous_position = position

print(f"The password is {password}.")
```

## Part 2

```python
password = 0
previous_position = starting_position
for rotation in dial_rotations:
    position = (previous_position + rotation) % 100
    crossed_zeros = abs((previous_position + rotation) // 100)

    # When the rotation is negative, make sure that the correct rotations are accounted for
    if rotation < 0:
        if position == 0:
            crossed_zeros += 1
        if previous_position == 0:
            crossed_zeros -= 1
    
    password += crossed_zeros
        
    previous_position = position

print(f"The actually correct password is {password}.")
```
